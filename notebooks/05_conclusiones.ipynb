Inspección inicial

Durante la inspección inicial del dataset se realizó una exploración general de la estructura de los datos utilizando funciones como head(), info() y describe(). Esto permitió identificar el tipo de variables, la distribución general de los datos y la presencia de posibles inconsistencias.

Se detectaron valores atípicos en variables numéricas como edad y tiempo de uso, así como diferencias en la escritura de categorías como países y planes de suscripción. Estos problemas fueron clave para definir las etapas posteriores de limpieza.

  Limpieza de datos

En la etapa de limpieza se corrigieron inconsistencias detectadas previamente. Se estandarizaron categorías como países y planes de suscripción, unificando valores escritos de diferentes formas.

También se trataron valores atípicos en variables numéricas, como edades fuera de rango lógico y tiempos de uso excesivos, reemplazándolos o filtrándolos para evitar que afecten el análisis.

Este proceso permitió mejorar la calidad del dataset y asegurar mayor confiabilidad en los resultados.
  Análisis Exploratorio de Datos (EDA)

El análisis exploratorio permitió identificar patrones en el comportamiento de los usuarios. Se analizaron distribuciones de variables como edad, país, género favorito y plan de suscripción.

A través de gráficos y medidas estadísticas se observaron tendencias generales de consumo y diferencias entre grupos de usuarios. El EDA fue fundamental para comprender la estructura del dataset y preparar la información para análisis más avanzados.

 PCA (Análisis de Componentes Principales)

El PCA fue aplicado para reducir la dimensionalidad del dataset utilizando variables numéricas estandarizadas. Esto permitió representar la información en un espacio más simple mediante componentes principales.

Se observó que la mayor parte de la varianza se concentra en el primer componente, lo que indica que resume gran parte de la información original. Esto facilita la visualización de patrones generales en los datos.

Limitaciones

El alcance de las conclusiones se encuentra condicionado por la información disponible en el dataset y por las decisiones tomadas durante el proceso de limpieza y transformación de los datos.
Asimismo, el análisis se basa únicamente en las variables proporcionadas, sin incluir información externa que podría enriquecer los resultados o aportar mayor profundidad al estudio.

 Mejoras futuras

Una mejora futura podría consistir en incorporar nuevas variables o fuentes de datos adicionales que permitan ampliar el alcance del análisis.
También se podría aplicar modelos de machine learning más avanzados para predecir comportamientos de los usuarios, así como optimizar el proceso de limpieza mediante técnicas automáticas de detección de outliers y valores inconsistentes.

 CONCLUSIÓN FINAL

El presente trabajo permitió desarrollar un flujo completo de análisis de datos, abarcando desde la inspección inicial del dataset hasta la aplicación de técnicas de análisis exploratorio y reducción de dimensionalidad mediante PCA.

A lo largo del proceso se logró limpiar, estructurar y estandarizar la información, lo que permitió mejorar significativamente la calidad del dataset y garantizar la confiabilidad de los resultados obtenidos.

Asimismo, el análisis realizado permitió identificar patrones de comportamiento relevantes en los usuarios y simplificar la complejidad de los datos para facilitar su interpretación.
