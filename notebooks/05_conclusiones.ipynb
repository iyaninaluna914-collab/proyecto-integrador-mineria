CONCLUSIONES
El proyecto integrador consistió en el análisis de un dataset de usuarios de una plataforma de streaming, aplicando técnicas de minería de datos.

Se desarrolló un proceso completo que incluye inspección inicial, limpieza de datos, análisis exploratorio (EDA) y reducción de dimensionalidad (PCA).

Durante la inspección inicial se identificaron valores nulos, registros duplicados e inconsistencias en variables categóricas.

Estas anomalías fueron tratadas mediante eliminación de duplicados, imputación de valores nulos y normalización de categorías.

El análisis exploratorio permitió identificar patrones en el comportamiento de los usuarios.

Se observaron diferencias en el tiempo de uso según edad y género de contenido favorito, así como variaciones en los planes de suscripción.

También se identificaron relaciones entre variables numéricas mediante análisis de correlación.

La técnica de PCA permitió reducir la dimensionalidad del dataset a dos componentes principales.

Esto facilitó la visualización de los datos y la identificación de patrones generales sin pérdida significativa de información.

Las variables numéricas seleccionadas contienen información suficiente para representar el comportamiento general de los usuarios.

El análisis está limitado al dataset provisto por la cátedra, sin incorporación de fuentes externas.

Además, el estudio no incluye modelos predictivos, ya que el alcance del proyecto se centra en análisis exploratorio y reducción de dimensionalidad.

Como mejora futura, se podrían incorporar nuevos datasets para enriquecer el análisis.

También sería posible aplicar modelos de machine learning para predicción del comportamiento de usuarios o segmentación avanzada.

El proyecto permitió aplicar un flujo completo de análisis de datos, desde la limpieza hasta la interpretación final.

Se logró transformar un dataset crudo en información estructurada y comprensible, utilizando técnicas fundamentales de Minería de Datos 1.

Los resultados obtenidos permiten comprender el comportamiento general de los usuarios y sirven como base para futuros análisis más avanzados.
